#3.1.1: Matplotlib Basics

```python
import pandas as pd
import numpy as np
```

```python
#import matplotlib
import matplotlib.pyplot as plt
```

```python
x = [0,1,2]
y = [100,200,300]
plt.plot(x,y)
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                        'price':[100,120,190,200,230,310,330,305]})
plt.scatter(housing['rooms'],housing['price'])
```

```python
# Adding labels, changing graph x and y axis boundaries + aesthetical customisation
plt.plot(x,y,color='red', marker='o',markersize=15,linestyle='--')

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

#3.1.2: Seaborn Basics

```python
#importing seaborn
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/Work 2022/Coding/Tensorflow/DATA/heart.csv')
```

```python
df.head()
```

```python
# distribution plot, without the kde line, increase size of plot, change colour
plt.figure(figsize=(12,8))
sns.distplot(df['age'],kde=False,color='red')
```

```python
#seaborn countplot, further segment data using hue
sns.countplot(x='cp',data=df,hue='sex',palette='terrain')
```

```python
#boxplot
sns.boxplot(x='target',y='thalach',data=df,hue='sex')
```

```python
# Edit size of points using size attribute
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex',palette='Dark2',size='age')
```

```python
diamonds = pd.read_csv('/content/drive/MyDrive/Work 2022/Coding/Tensorflow/DATA/diamonds.csv')
```

```python
diamonds.info()
```

```python
#Scatterplot of price versus carat
plt.scatter(diamonds['carat'],diamonds['price'])
```

```python
plt.scatter(diamonds['carat'],diamonds['price'],alpha=0.1,edgecolors='none')
```

```python
plt.figure(figsize=(12,8))
plt.scatter(diamonds['carat'],diamonds['price'],alpha=0.1,edgecolors='none')
```

```python
sns.countplot(diamonds['cut'])
```

```python
sns.boxplot(x='cut',y='price',data=diamonds)
```

```python
sns.boxplot(x='cut',y='price',order=['Fair','Ideal','Good','Very Good','Premium'],data=diamonds)
```
